```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Create two example data sets

```python
import numpy as np
random_walk =np.cumsum(np.random.randn(1000,3),axis=0)
tips = sns.load_dataset("tips")
```

Plot a regular matplolib line plot

```python
ax = plt.plot(random_walk)
```

With a Seaborn palplot one can plot the current matplotlib/Seaborn color cycle.

```python
sns.palplot(plt.rcParams['axes.prop_cycle'].by_key()['color'])
```

The popular stats library Seaborn uses its own color palette per default.

```python
ax = sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips)
plt.savefig("regular.png")
```

In this example notebook, we use the installed correlaidmatplotlib package to style the plots. The import already changes the style to the CorrelAid theme. There is no need to place the style file in the user folder, and no need to use `style.use(...)`. Again, the style is applied to the matplotlib and Seaborn plots, and the color cycler is used.

```python
import correlaidmatplotlib

ax = plt.plot(random_walk)
```

```python
sns.palplot(sns.color_palette())
```

```python
sns.barplot(x="day", y="total_bill", hue="smoker", data=tips)
```

While the default Seaborn plots are quite powerful and the style already improves some defaults, there is still room for improvement. With the CorrelAid package we provide helper functions to further style the plots consistently, i.e., we:
- add Title and subtitle,
- add a signature bar adding a Copyright and data source,
- take away the x and y axis labels, that are not necessary, when using the subtitle,
- move the legend to the right of the figure

```python
from correlaidmatplotlib import set_correlaid_style

fig, ax = plt.subplots()
sns.barplot(x="day", y="total_bill", hue="smoker", data=tips, ax=ax)
set_correlaid_style(fig, "Tipping Habits of Smokers", "Bill total, in US Dollar", source="tips.csv")
```

```python
fig, ax = plt.subplots()
lines = ax.plot(random_walk)
set_correlaid_style(fig, "Comparision of Three Random Walks", "Random Walk, some random domain", source="np.random")
```

We provide individual functions to control also individual elements. 

```python
from correlaidmatplotlib import add_title_and_subtitle
fig, ax = plt.subplots()
sns.barplot(x="day", y="total_bill", hue="smoker", data=tips, ax=ax)
add_title_and_subtitle(fig, "Tipping Habits of Smokers", "Bill total, in US Dollar")
```

```python
from correlaidmatplotlib import add_signature
fig, ax = plt.subplots()
sns.barplot(x="day", y="total_bill", hue="smoker", data=tips, ax=ax)
add_signature(fig, "tips.csv", fontsize='small')
```

```python
from correlaidmatplotlib import remove_axes_labels
fig, ax = plt.subplots()
sns.barplot(x="day", y="total_bill", hue="smoker", data=tips, ax=ax)
remove_axes_labels(fig)
```

```python
from correlaidmatplotlib import place_legend_outside
fig, ax = plt.subplots()
sns.barplot(x="day", y="total_bill", hue="smoker", data=tips, ax=ax)
place_legend_outside(ax)
```

The default placements and fontsizes may not be ideal for each plot. Therefore, they can be adjusted with function.

```python
fig, ax = plt.subplots()
sns.barplot(x="day", y="total_bill", hue="smoker", data=tips, ax=ax)
set_correlaid_style(fig, "Tipping Habits of Smokers", "Bill total, in US Dollar", source="tips.csv",
                   title_x_pos=0.03, subtitle_x_pos=0.03, copyright_x_pos=0.03,
                   title_y_pos=0.1, subtitle_y_pos=0.0, signature_fontsize="smaller")
```
