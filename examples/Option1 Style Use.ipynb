```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Create two example data sets

```python
import numpy as np
random_walk =np.cumsum(np.random.randn(1000,3),axis=0)
tips = sns.load_dataset("tips")
```

Plot a regular matplolib line plot

```python
ax = plt.plot(random_walk)
```

With a Seaborn palplot one can plot the current matplotlib/Seaborn color cycle.

```python
sns.palplot(plt.rcParams['axes.prop_cycle'].by_key()['color'])
```

The popular stats library Seaborn uses its own color palette per default.

```python
ax = sns.barplot(x="day", y="total_bill", hue="smoker", data=tips)
```

Now that we have viewed the default options, we change the style to the correlaid style.

```python
import matplotlib.style as style
style.use("correlaid")
```

Now, the default plot uses the CorrelAid colors.

```python
ax = plt.plot(random_walk)
```

The color palette shows now th CorrelAid scheme colors.

```python
sns.palplot(sns.color_palette())
```

The color scheme is now also recognized by the Seaborn library for statistical plots.

```python
ax = sns.barplot(x="day", y="total_bill", hue="smoker", data=tips)
```
